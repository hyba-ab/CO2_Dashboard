```python
import pandas as pd 
import numpy as np 
import panel as pn 
pn.extension('tabulator')

import hvplot.pandas
```

```python
df=pd.read_csv('owid-co2-data.csv')
```

```python
df.columns
```

```python
df[df['country']=='World']
```

```python
## Some data preprocessing 
```

```python
# fill NAs with 0s and add GDP per capita column 
```

```python
df=df.fillna(0)
df['gdp_per_capita']=np.where(df['population']!=0 , df['population'],0)
```

```python
df
```

```python
## making the pipeline interactive 
```

```python
idf=df.interactive()
```

```python
# defining panel widgets 
```

```python
Year_slider =pn.widgets.IntSlider(name='Year slider',start=1750,end=2022,step=2,value=1850)
```

```python
Year_slider
```

```python
# creating radio buttons for co2 measures
```

```python
yaxis_co2=pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['co2','co2_per_capita',],
    button_type='success'
)
```

```python
Year_slider =pn.widgets.IntSlider(name='Year slider',start=1750,end=2022,step=2,value=1850)
continents = ['World', 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_pipeline = (
    idf[
        (idf['year'] <= Year_slider) &  
        (idf['country'].isin(continents))
    ]
    .groupby(['country', 'year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
)


print(co2_pipeline)
```

```python
co2_plot=co2_pipeline.hvplot(x='year', by='country',y=yaxis_co2,line_width=2,title="CO2 Emission by Continent")
co2_plot
```

```python
## Table : co2 emission by continent over time 
```

```python
co2_table=co2_pipeline.pipe(pn.widgets.Tabulator,pagination='remote',page_size=10,sizing_mode='stretch_width')
co2_table
```

```python
## Scatteplot : CO2 VS GDP 
```

```python
co2_vs_gdp_scatterplot_pipeline=(
    idf[
    (idf.year==Year_slider) &
    (~(idf.country.isin(continents)))
    ]
    .groupby(['country', 'year','gdp_per_capita'])['co2'].mean(numeric_only=True) 
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
)
```

```python
co2_vs_gdp_scatterplot=co2_vs_gdp_scatterplot_pipeline.hvplot(x='gdp_per_capita',y='co2',by='country',size=80,kind="scatter",alpha=0.7,legend=False,height=500,width=500)
```

```python
co2_vs_gdp_scatterplot
```

```python
## Bar chart : co2 sources by continent 
```

```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['coal_co2','oil_co2','gas_co2'],
    button_type='success'
)


continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']


co2_source_bar_pipeline=(
    idf[
    (idf.year==Year_slider) &
    (~(idf.country.isin(continents_excl_world)))
    ]
    .groupby(['country', 'year'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
)
```

```python
co2_source_bar_plot=co2_source_bar_pipeline.hvplot(kind='bar',x='country',y=yaxis_co2_source,title='CO2 Source by Continent')

co2_source_bar_plot
```

```python
## preparing the dashboard 
```

```python
#Layout using Template
template = pn.template. FastListTemplate(
    title='World CO2 emission dashboard',
    sidebar=[pn.pane.Markdown("# C02 Emissions and Climate Change"),
             pn.pane.Markdown("#### Carbon dioxide emissions are the primary driver of global climate change."),
             pn.pane.PNG('climate_day.png', sizing_mode='scale_both'),
             pn.pane.Markdown("## Settings"),
             Year_slider],
    main=[pn.Row(pn.Column (yaxis_co2, co2_plot.panel (width=700), margin=(0,25)), co2_table. panel (width=500)),
          pn.Row(pn.Column (co2_vs_gdp_scatterplot.panel (width=600), margin=(0,25)), 
                 pn.Column (yaxis_co2_source, co2_source_bar_plot.panel (width=600)))],
)
accent_base_color="#88d8b0",
header_background="#88d8b0",
# template.show()
template.servable();
```
